```python
import math

# Haversine function to calculate the distance between two points on the Earth
def haversine(lon1, lat1, lon2, lat2):
    # Convert latitude and longitude from degrees to radians
    lon1, lat1, lon2, lat2 = map(math.radians, [lon1, lat1, lon2, lat2])
    
    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    # Radius of Earth in kilometers (mean value)
    r = 6371.0
    
    # Calculate the distance
    return c * r

# Example proximity point (e.g., Kennedy Space Center)
proximity_location = (28.5733, -80.6469)

# Iterate through the launch sites and calculate the distance
for index, record in launch_sites_df.iterrows():
    launch_site_location = (record['Lat'], record['Long'])
    distance = haversine(launch_site_location[1], launch_site_location[0], proximity_location[1], proximity_location[0])
    
    print(f"Distance from {record['Launch Site']} to proximity location: {distance:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # Approximate radius of Earth in km
    R = 6371.0

    # Convert latitude and longitude from degrees to radians
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    # Calculate differences
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # Calculate distance
    distance = R * c
    return distance

# Example usage
launch_site_lat = 28.56367
launch_site_lon = -80.57163
coastline_lat = 28.56367  # Replace with actual coastline latitude
coastline_lon = -80.57163  # Replace with actual coastline longitude

distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance from launch site to coastline: {distance_coastline:.2f} km")
```

```python
import folium
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Function to calculate distance using the Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # Radius of Earth in kilometers

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Initialize the map
site_map = folium.Map(location=[28.5623, -80.5774], zoom_start=5)

# Example coordinates for the launch site and a point of interest (city, railway, highway)
launch_site_lat = 28.56367
launch_site_lon = -80.57163
point_of_interest_lat = 28.6142  # Example: City or railway location
point_of_interest_lon = -80.6480

# Calculate distance from the launch site to the point of interest
distance = calculate_distance(launch_site_lat, launch_site_lon, point_of_interest_lat, point_of_interest_lon)

# Create a marker for the point of interest with a DivIcon showing the distance
distance_marker = folium.Marker(
    location=[point_of_interest_lat, point_of_interest_lon],
    icon=DivIcon(
        icon_size=(200, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance),
    )
)

# Add the marker to the map
site_map.add_child(distance_marker)

# Create and add a marker for the launch site
launch_marker = folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup="Launch Site",
    icon=folium.Icon(color='blue', icon='rocket')
)
site_map.add_child(launch_marker)

# Draw a line between the launch site and the point of interest
line = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [point_of_interest_lat, point_of_interest_lon]],
    color='blue',
    weight=2.5,
    opacity=1
)

# Add the line to the map
site_map.add_child(line)

# Display the map
site_map
```
